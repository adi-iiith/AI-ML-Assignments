<h2> Part 1.1.1 </h2>

```python
alist = [1,2,3,4,5,6,7,8,9]
file = 'Datasets/Robot1'
df = pd.read_csv(file,sep=" ",names=alist)
df = df[[2,3,4,5,6,7,8,9]]
train = df.sample(frac=0.8,random_state=200)
validate = df.drop(train.index)
alist.remove(1)
alist.remove(2)
alist.remove(9)
# df.info()
# print(df)

label = df[2].unique().tolist()

def eucledian_distance(x,y):
    distance = 0
    for i in alist:
        distance += pow((x[i]-y[i]),2)
    return math.sqrt(distance)

def neighbours(df,instance,k):
    distance = []
    for i,row in train.iterrows():
        x = eucledian_distance(row,instance)
        temp = (x,row[2])
        distance.append(temp)
    first_k = sorted(distance,key = lambda t:t[0])
    return first_k[:k]

def response(train,test,k):
    distance = []
    for i,row in test.iterrows():
        distance.append(neighbours(train,row,k))
    return distance

k_distance = response(train,validate,1)

prediction = []
for i in k_distance:
    iris_count = [0,0]
    for x in i:
        if x[1] == label[0]:
            iris_count[0]+=1
        elif x[1] == label[1]:
            iris_count[1]+=1

    temp = iris_count.index(max(iris_count))
    prediction.append(label[temp])

actual = validate[2].tolist()
accuracy = 0
for i in range(len(actual)):
    if(prediction[i]==actual[i]):
        accuracy+=1
print("Accuracy : ",accuracy/len(actual))

def predict(k_distance):
    prediction = []
    for i in k_distance:
        iris_count = [0,0,0]
    #     ['Iris-setosa', 'Iris-virginica', 'Iris-versicolor']
        for x in i:
            if x[1] == label[0]:
                iris_count[0]+=1
            elif x[1] == label[1]:
                iris_count[1]+=1
            else:
                iris_count[2]+=1

        temp = iris_count.index(max(iris_count))
        prediction.append(label[temp])
    return prediction

def find_accuracy(k):
    k_distance = response(train,validate,k)
    prediction = predict(k_distance)
    actual = validate[2].tolist()
    accuracy = 0
    for i in range(len(actual)):
        if(prediction[i]==actual[i]):
            accuracy+=1
    print(k,"::: ",accuracy/len(actual))

for i in range(1,15):
    find_accuracy(i)
```

<h2>Part 2</h2>

```python
fname = 'Datasets/data_part2.csv'
alist = ['ID','Age','Experience','Annual_Income','Zipcode','Family_size',
         'Monthly_spend','Education','Mortgage','target','Security','CD_account','Net_Banking','Ccard']
df = pd.read_csv(fname,names = alist)
df = df.iloc[1:]
ndata = ['Age','Experience','Annual_Income','Monthly_spend','Mortgage']
cdata = list(set(alist)-set(ndata))
cdata.remove('target')
cdata.remove('ID')
cdata.remove('Zipcode')

# df = df[df.Experience != -1]
# a=df.loc[df['Experience']==-1].index

# print(a)
cdata

train = df.sample(frac=0.8,random_state=200)
validate = df.drop(train.index)

# computation for numeric data
mean = {}
deviation = {}
label = ['Yes' , 'No']
for i in ndata:
    mean[i] = {}
    deviation[i] = {}
#     x = train[train['target']==1][i].sum()
#     len = train[train['target']==1][i].count()
    mean[i][label[0]] = train[train['target']==1][i].mean()
    mean[i][label[1]] = train[train['target']==0][i].mean()
    deviation[i][label[0]] = train[train['target']==1][i].std()
    deviation[i][label[1]] = train[train['target']==0][i].std()

# computation for categorical data
category = {}
for i in cdata:
    category[i] = {}
    uni = train[i].unique()
    for val in uni:
        category[i][val] = {}
        x = train[train[i] == val]
        category[i][val][label[0]] = x[x['target']==1][i].count() / train[train['target']==1][i].count()
        category[i][val][label[1]] = x[x['target']==0][i].count() / train[train['target']==0][i].count()

def normal(mean,variance,w):
    dem = (math.sqrt(2 * math.pi)*variance)
    num = math.exp(-1 * (pow(((w-mean)/variance),2)) )
    return (num/dem)
    

# Prediction

y = train[train['target']==1]['Age'].count()
n = train[train['target']==0]['Age'].count()
p_yes = y/(y+n)
p_no = n/(y+n)

prediction = []

for i,row in validate.iterrows():
    yes , no = 1,1
    for j in ndata:
        yes *= normal(mean[j]['Yes'],deviation[j]['Yes'],row[j])
        no *= normal(mean[j]['No'],deviation[j]['No'],row[j])
    for j in cdata:
        yes = yes * category[j][row[j]]['Yes']
        no = no * category[j][row[j]]['No']
    
    yes *= p_yes
    no *= p_no
    if(yes >= no):
        prediction.append(1)
    else:
        prediction.append(0)

# accuracy

accuracy = 0 
actual = validate['target'].tolist()
for i in range(len(actual)):
    if(actual[i] == prediction[i]):
        accuracy += 1

accuracy = accuracy/len(actual)
print(accuracy)
```

<h2>Part 3</h2>

```python
fname = 'Datasets/data_part3.csv'
df = pd.read_csv(fname)
df = df.drop('Serial No.',axis=1)
train = df.sample(frac=0.8,random_state=200)
validate = df.drop(train.index)

# print(df)

y = train['Chance of Admit ']
train = train.drop('Chance of Admit ',axis=1)

actual = validate['Chance of Admit ']
validate = validate.drop('Chance of Admit ',axis=1)


# adding 1's
one = []
for i in range(len(train.index)) :
    one.append(1)
train.insert(0, 'ones', one, allow_duplicates = False)
# Dataframe to Matrix

# Validation data handling
one_validate = []
for i in range(len(validate.index)) :
    one_validate.append(1)
validate.insert(0, 'ones', one_validate, allow_duplicates = False)

X = train.values
Y = y.values


Xt = X.transpose()
XXt = np.matmul(Xt,X)
X_inv = np.linalg.inv(XXt)


B = np.matmul((np.matmul(X_inv,Xt)),Y)

X_validate = validate.values

prediction = np.matmul(X_validate,B)
prediction = prediction.tolist()



actual = actual.tolist()

error_count = 0

for i in range(len(actual)) :
    error_count = error_count + (pow((actual[i] - prediction[i]),2) * 0.5)
print("MSE :: ",error_count)


MAE = 0
for i in range(len(actual)) :
    MAE = MAE + (abs((actual[i] - prediction[i])))
MAE = MAE/len(actual)
print("MAE :: ",MAE)


MAPE = 0
for i in range(len(actual)) :
    MAPE = MAPE + (abs((actual[i] - prediction[i])/actual[i]))
MAPE = 100 * MAPE/len(actual)
print("MAPE :: ",MAPE,"%")
```
