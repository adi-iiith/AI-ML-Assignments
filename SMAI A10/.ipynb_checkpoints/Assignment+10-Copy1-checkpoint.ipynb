```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
train = df[100:]
test = df[:100]
# train = X[132:]
# test = X[:132]
```

```python
len(train) + len(test)
```

```python
# train = pd.concat([X1_train,y1_train], axis=1)
# test = pd.concat([X1_test,y1_test], axis=1) 
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
train = sc.fit_transform(train)
test = sc.transform(test)
```

```python
train
```

```python
X = train[:,0:2]
# X = train[:,0:1]
X
```

```python
X.shape
```

```python
Y = train[:,2]
# Y = train[:,1]
```

```python
Y.shape
```

```python
time_step = [20,50,75]
```

```python
for ts in time_step: 
    X_train = []
    y_train = []
    for i in range(ts,X.shape[0]):
        X_train.append(X[i-ts:i])
        y_train.append(Y[i])
    X_train, y_train = np.array(X_train), np.array(y_train)
    y_train = y_train.reshape(-1,1)
    # X_train = scx.fit_transform(X_train)
    # y_train = scy.fit_transform(y_train)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 2))


    for hn in [30,50,80]:
    
        regressor = Sequential()
        # Adding the first LSTM layer and some Dropout regularisation
        regressor.add(LSTM(units = hn, return_sequences = True, input_shape = (X_train.shape[1], 2)))
        regressor.add(Dropout(0.2))

        # Adding a second LSTM layer and some Dropout regularisation
        regressor.add(LSTM(units = hn, return_sequences = True))
        regressor.add(Dropout(0.2))

        # Adding a third LSTM layer and some Dropout regularisation
        regressor.add(LSTM(units = hn))
        regressor.add(Dropout(0.2))

        # Adding the output layer
        regressor.add(Dense(units = 1))
        # Compiling the RNN
        regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

        regressor.fit(X_train, y_train, epochs = 100, batch_size = 32)


        Xt = test[:,0:2]
        Yt = test[ts:,2]


        X_test = []
        y_test = []
        for i in range(ts,Xt.shape[0]):
            X_test.append(Xt[i-ts:i])
        #     y_test.append(Yt[i])
        X_test, y_test = np.array(X_test), np.array(y_test)
        y_test = y_test.reshape(-1,1)
        # y_test = sc.transform(y_test)

        p = regressor.predict(X_test)

        from sklearn.metrics import mean_squared_error as mse
        import math
        print("::----------MSE-----------::")
        print(math.sqrt(mse(p,Yt)))
    for hn in [30,50,80]:
        regressor = Sequential()
        # Adding the first LSTM layer and some Dropout regularisation
        regressor.add(LSTM(units = hn, return_sequences = True, input_shape = (X_train.shape[1], 2)))
        regressor.add(Dropout(0.2))

        # Adding a second LSTM layer and some Dropout regularisation
        regressor.add(LSTM(units = hn))
        regressor.add(Dropout(0.2))

        # Adding the output layer
        regressor.add(Dense(units = 1))
        # Compiling the RNN
        regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')

        regressor.fit(X_train, y_train, epochs = 100, batch_size = 32)


        Xt = test[:,0:2]
        Yt = test[ts:,2]


        X_test = []
        y_test = []
        for i in range(ts,Xt.shape[0]):
            X_test.append(Xt[i-ts:i])
        #     y_test.append(Yt[i])
        X_test, y_test = np.array(X_test), np.array(y_test)
        y_test = y_test.reshape(-1,1)
        # y_test = sc.transform(y_test)

        p = regressor.predict(X_test)

        from sklearn.metrics import mean_squared_error as mse
        import math
        print("::----------MSE-----------::")
        print(math.sqrt(mse(p,Yt)))
```
